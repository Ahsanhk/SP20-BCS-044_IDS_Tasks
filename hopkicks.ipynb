```python
from bs4 import BeautifulSoup as Soup
import requests
import pandas as pd
import json
```

```python
url = "https://www.hopkicks.pk/Nike" 
resp = requests.get(url) 
resp = resp.text 
```

```python
soup = Soup(resp, 'html.parser')
```

```python
divs= soup.findAll('div', {'class':'product-thumb'})
```

```python
divs
```

```python
products = []
for div in divs:
    name = div.find('div', {'class':'caption'}).text.strip()
    price = div.find('p', {'class': 'price'}).text.strip()
    image = div.find('img', {'class':'img-responsive'})
    jsonObj = {"name":name, "price": price, "image": image}
    products.append(jsonObj)
```

```python
products
```

```python
df = pd.DataFrame(products)
df
```
