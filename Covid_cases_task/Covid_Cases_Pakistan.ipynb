```python
import folium
import pandas as pd


map_pak = folium.Map(location = [31.17, 72.70], zoom_start = 6.5, tiles = 'OpenStreetMap')

df = pd.read_csv('./pak_covid_data.csv')
df
```

```python
pt = 0
for cases in df['Total Cases']:
    pt = int(cases) + int(pt)

for state, lat, long, ct, agr, dr in zip(list(df['City']), list(df['Latitude']), list(df['Longitude']), list(df['Total Cases']), list(df['Avg Growth Rate']), list(df['Doubling Rate (Days)'])):
    folium.CircleMarker(location = [lat, long], radius = (ct/pt)*100, color = 'red', fill = False, fill_color = 'blue').add_to(map_pak)
    html_format = '<strong>City:  ' + state + '</strong><br>' + 'Total Cases: ' + str(ct) + '<br>' + 'Avg Growth Rate: '+ str(agr) + '<br>' + 'Doubling Rate (Days): ' + str(dr)
    iframe = folium.IFrame(html_format, width = 250, height = 90)
    popupData = folium.Popup(iframe, max_width = 500)
    folium.Marker(location = [lat, long], popup = popupData).add_to(map_pak)
    
map_pak
```
