```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
str1 = 'sunshine state enjoy sunshine'
str2 = 'brown fox jump high, brown fox run'
str3 = 'sunshine state fox run fast'

corpus = [str1, str2, str3]
corpus
```

```python
bow_vectorizer = CountVectorizer()
bow = bow_vectorizer.fit_transform(corpus)

features = bow_vectorizer.get_feature_names()
vocabulary = bow_vectorizer.vocabulary_
vectors = bow.toarray()

i = 0
for i in range(0, len(vectors)):
    print('string', i, ':', vectors[i])
    i =+ 1
```

```python
bow = pd.DataFrame(bow.toarray(), columns = features)
bow
```

```python
tfidf_vectorizer = TfidfTransformer()
var = tfidf_vectorizer.fit_transform(bow)
idf = pd.DataFrame({'feature':features, 'idf':tfidf_vectorizer.idf_})
idf
```

```python
tf_idf = pd.DataFrame(np.round(var.toarray(),3) ,columns=features)
tf_idf
```

```python
similarity = cosine_similarity(bow.iloc[[0,2]])

result = pd.DataFrame(similarity)
result
```
